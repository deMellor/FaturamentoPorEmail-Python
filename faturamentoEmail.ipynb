```python
import pyautogui   #automatizar teclado e mouse
import pyperclip   #control c e control v
import time        #biblioteca de tempo

pyautogui.FAILSAFE = True
pyautogui.PAUSE = 2

# Passo 1 - Entrar no sistema do drive
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/1TvE5lCT_EDn9TInZhIWKUJVafd0c7h15?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# Passo 2 - Entrar na base de dados
time.sleep(5)
pyautogui.click(x=411, y=308, clicks=2)

# Passo 3 - Exportar a base de vendas
time.sleep(5)
pyautogui.click(x=411, y=308)
time.sleep(2)
pyautogui.click(x=1157, y=186)
time.sleep(2)
pyautogui.click(x=905, y=595)
time.sleep(5)
```

```python
# Passo 4 - Calcular indicadores (faturamento e quantidade de produtos vendidos)
import pandas as pd

tabelaVendas = pd.read_excel(r"C:\Users\user\Downloads\Vendas - Dez.xlsx")
display(tabelaVendas)

faturamento = tabelaVendas["Valor Final"].sum()
quantidade = tabelaVendas["Quantidade"].sum()

```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5 - Enviar um email para a diretoria com os indicadores
pyautogui.PAUSE = 1


pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(3)

#clicar em novo email
pyautogui.click(x=122, y=189)
time.sleep(5)

#escrever destinatario
pyautogui.write("marcosmello2022r@outlook.com")
pyautogui.press("tab")

#escrever assunto
pyautogui.press("tab")
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")

#escrever corpo do email
pyautogui.press("tab")
textoEmail = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abraços
BotPython
"""

pyperclip.copy(textoEmail)
pyautogui.hotkey("ctrl", "v")

#enviar email ctrl enter
pyautogui.hotkey("ctrl", "enter")
```
